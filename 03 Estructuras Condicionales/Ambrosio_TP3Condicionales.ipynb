```python
#1. Escribir un programa que solicite la edad del usuario. Si el usuario es mayor de 18 años,
#deberá mostrar un mensaje en pantalla que diga “Es mayor de edad”.

#Inicialización de variable con el valor ingresado por el usuario
edad=int(input("Ingrese su edad: "))

#Estructura de control: determinar mayoría de edad
#Condicional Ternario
print("Es mayor de edad") if edad >= 18 else print("No es mayor de edad")

#Condicional simple
if edad >= 18:
    print("Es mayor de edad")
else:
    print("No es mayor de edad")
```

```python
#4 Escribir un programa que solicite al usuario su edad e imprima por pantalla a cuál de las
#siguientes categorías pertenece:
#   ● Niño/a: menor de 12 años. (edad < 12)
#   ● Adolescente: mayor o igual que 12 años y menor que 18 años. (edad >= 12 and edad < 18)
#   ● Adulto/a joven: mayor o igual que 18 años y menor que 30 años. (edad >= 18  and edad < 30)
#   ● Adulto/a: mayor o igual que 30 años. (edad >= 30)

#Solicitar edad
edad=int(input("Ingrese su edad: "))

#Condicional compuesto1: Identificar categoría
if edad < 12:
    print("Niño/a")
elif (edad >= 12 and edad < 18):
    print("Adolescente")
elif (edad >= 18  and edad < 30):
    print("Adulto/a joven")
else:
    print("Adulto/a mayor")

#Condicional compuesto2: Identificar categoría
if edad < 12:
    print("Niño/a")
elif edad < 18:
    print("Adolescente")
elif edad < 30:
    print("Adulto/a joven")
else:
    print("Adulto/a mayor")

#Condicional anidado: Identificar categoría

if edad >= 12:
    if edad < 18:
        print("Adolescente")
    else:
        if edad >= 30:
            print("Adulto/a mayor")  
        else:
            print("Adulto/a joven")
else:
    print("Niño/a")
```

```python
#5 Uso de la función Len()
#The Python len() function is used to return a numeric value that denotes the length of the given list,
# tuple, string, array, dictionary, etc.

# Escribir un programa que permita introducir contraseñas de entre 8 y 14 caracteres (incluyendo 8 y 14). 
# Si longitud adecuada, imprimir "Ha ingresado una contraseña correcta";
# en caso contrario, imprimir "Por favor, ingrese una contraseña de entre 8 y 14 caracteres".


#Solicitar valor al usuario
password=input("Escriba una contraseña de entre 8 y 14 dígitos: ")
passLength= len(password)

#Corroborar: condicion longitud de contraseña

if passLength >= 8 and passLength <= 14:
    print("Ha ingresado una contraseña correcta")
else:
    print("Por favor, ingrese una contraseña de entre 8 y 14 caracteres")

#Condicional terciario
print("Ha ingresado una contraseña correcta") if (passLength >= 8 and passLength <= 14) else print("Por favor, ingrese una contraseña de entre 8 y 14 caracteres")
```

```python
#6 Escribir un programa que tome la lista numeros_aleatorios, calcule su moda, su mediana y su media 
# y las compare para determinar si hay sesgo positivo, negativo o no hay sesgo. 
# Imprimir el resultado por pantalla.
#Utilizar la librería statistics
# numeros_aleatorios crea una lista con 50 números entre 1 y 100 elegidos de forma aleatoria.
#Sesgo positivo o a la derecha: media > mediana > moda
#Sesgo negativo o a la izquierda: media < mediana < moda
#Sin sesgo: media == mediana == moda

#Crea lista de números aleatorios 
import random
numeros_aletorios = [random.randint(1, 100) for i in range(50)]

#Importa funciones desde el módulo "statistics"
from statistics import mode, median, mean

#Define las variables que contendran los resultados de los cálculos
media = mean(numeros_aletorios)
moda = mode(numeros_aletorios)
mediana = median(numeros_aletorios)

#Condicional1: predecir distribucion

if media > mediana > moda:
    print("Sesgo positivo o a la derecha")
elif media < mediana < moda:
    print("Sesgo negativo o a la izquierda")
else:
    print("Sin sesgo")

#Condicional2: predecir distribucion

if media > mediana and media > moda:
    print("Sesgo positivo o a la derecha")
elif media < mediana and media < moda:
    print("Sesgo negativo o a la izquierda")
else:
    print("Sin sesgo")

```

```python
# 9 Escribir un programa que pida al usuario la magnitud de un terremoto, 
# clasifiquela según la escala de Richter e imprima el resultado
#● Menor que 3: "Muy leve" (imperceptible).
#● Mayor o igual que 3 y menor que 4: "Leve" (ligeramente perceptible).
#● Mayor o igual que 4 y menor que 5: "Moderado" (sentido por personas, pero generalmente no causa daños).
#● Mayor o igual que 5 y menor que 6: "Fuerte" (puede causar daños en estructuras débiles).
#● Mayor o igual que 6 y menor que 7: "Muy Fuerte" (puede causar daños significativos).
#● Mayor o igual que 7: "Extremo" (puede causar graves daños a gran escala)

#Solicita un valor de magnitud y lo asigna a una variable
magnitud = int(input("Ingrese el valor de magnitud del terremoto (1 a 9)"))

#Solucion
if magnitud < 3:
    print("Muy Leve: imperceptible")
elif magnitud < 4:
    print("Leve: ligeramente perceptible")
elif magnitud < 5:
    print("Moderado: sentido por personas, pero generalmente no causa daños")
elif magnitud < 6:
    print("Fuerte: puede causar daños en estructuras débiles")   
elif magnitud < 7:
    print("Muy Fuerte: puede causar daños significativos") 
elif magnitud <= 9:
    print("Extremo: puede causar graves daños a gran escala")
else:
    print("ERROR: Ingrese un número entre 1 y 9")
```
